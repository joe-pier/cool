# dinamic_plotting

```python
class sierp():
    def __init__(self, iter, t_d = 50):
        self.fig = plt.figure()
        self.t_d = t_d  
        self.v1 = [random.randint(1,self.t_d),random.randint(1,self.t_d)]
        self.v2 = [random.randint(1,self.t_d),random.randint(1,self.t_d)]
        self.v3 = [random.randint(1,self.t_d),random.randint(1,self.t_d)]
        self.iter = iter
        self.vertex = [self.v1,self.v2,self.v3]
        self.initial_point = self.point_on_triangle()
        self.points = [self.v1,self.v2,self.v3, self.initial_point]
        self.r = []

    def random_point(self):
        return random.choice(self.vertex)

    def get_mid_point(self, p1,p2):
        return [(p1[0]+p2[0])/2, (p1[1]+p2[1])/2]

    def add_point(self):
        p = self.initial_point
        for i in range(0, self.iter):
            self.r = self.random_point()
            mid = self.get_mid_point(self.r, p)
            p = mid
            self.points.append(p)
            self.plot()
            plt.pause(0.1)
            plt.clf()
        return self.points

    def point_on_triangle(self):
        x, y = sorted([random.random(), random.random()])
        s, t, u = x, y - x, 1 - y
        return (s * self.v1[0] + t * self.v2[0] + u * self.v3[0],
                s * self.v1[1] + t * self.v2[1] + u * self.v3[1])

    def plot(self):
        ax = self.fig.add_subplot(111)
        v = list(zip(*self.vertex))
        ax.scatter(v[0],v[1],  marker='x', s = 70)
        ax.scatter(self.initial_point[0],self.initial_point[1], marker = ".", s = 150)
        t = plt.Polygon(self.vertex, alpha=0.1,edgecolor = "blue", linewidth=4)
        plt.gca().add_patch(t)
        for p in self.points[4:-1]:
            ax.scatter(p[0],p[1], marker = "o", s = 5,edgecolor = "black")
        ax.scatter(self.points[-1][0],self.points[-1][1], marker = "v", s = 200)
        ax.scatter(self.r[0],self.r[1], marker = "h", s = 200, alpha=0.5,edgecolor = "black")
```
